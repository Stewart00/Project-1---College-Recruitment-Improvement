```python
import pandas as pd
import matplotlib.pyplot as plt

data_2020 = pd.read_csv(r"C:\Users\stewa\Downloads\cleaned survey\Cleaned Majors Survey Results - Fall 2020.csv")
data_2021 = pd.read_csv(r"C:\Users\stewa\Downloads\cleaned survey\Cleaned Majors Survey Results - Fall 2021.csv")
data_2022 = pd.read_csv(r"C:\Users\stewa\Downloads\cleaned survey\Cleaned Majors Survey Results - Fall 2022.csv")
data_2023 = pd.read_csv(r"C:\Users\stewa\Downloads\cleaned survey\Cleaned Majors Survey Results - Fall 2023.csv")
data_2024 = pd.read_csv(r"C:\Users\stewa\Downloads\cleaned survey\Cleaned Majors Survey Results - Fall 2024.csv")
```

Clean data to plot

```python
def gender_counts(data):
    filtered_data = data[data['gender'] != 'Prefer not to say']
    return filtered_data['gender'].value_counts()

gender_2020 = gender_counts(data_2020)
gender_2021 = gender_counts(data_2021)
gender_2022 = gender_counts(data_2022)
gender_2023 = gender_counts(data_2023)
gender_2024 = gender_counts(data_2024)

gender_data = pd.DataFrame({
    '2020': gender_2020,
    '2021': gender_2021,
    '2022': gender_2022,
    '2023': gender_2023,
    '2024': gender_2024
}).T  # Transpose for better plotting

gender_data = gender_data.fillna(0)
```

Plotting the data

```python
colors = ['darkgrey', 'lightblue', 'grey', 'pink']
gender_data.plot(kind='bar', figsize=(10, 6), color=colors)
plt.title('Gender Demographics from 2020 to 2024')
plt.ylabel('Number of Respondents')
plt.xlabel('Year')
plt.xticks(rotation=0)  # Rotate x-axis labels for readability
plt.legend(title='Gender')
plt.show()
```

The graph shows a decline in total participants involving a steady decline for both males and females.
