Load file and get count of each age range

```python
import pandas as pd
import matplotlib.pyplot as plt

data_2024 = pd.read_csv(r"C:\Users\stewa\Downloads\cleaned survey\Cleaned Majors Survey Results - Fall 2024.csv", encoding='ISO-8859-1')

def age_range_counts(data):
    return data['age_'].value_counts()

age_2024 = age_range_counts(data_2024)
```

Plotting the pie chart using age data

```python
plt.figure(figsize=(8, 8))
plt.pie(age_2024, labels=age_2024.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'orange', 'yellow', 'pink', 'magenta'])
plt.title('Age Distribution of Students (2024)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
